```python
geneva_listings = pd.read_csv('../data/zurich/listings.csv')
geneva_listings.head()
```

```python
geneva_listings.columns
```

```python
grouped = geneva_listings.groupby('neighbourhood').mean()
grouped.head()
```

```python
grouped.columns
```

```python
df = grouped[['price', 'minimum_nights','number_of_reviews']]
df['minimum_nights']= df[['minimum_nights']].apply(lambda x: x.round(0))
df['number_of_reviews']= df[['number_of_reviews']].apply(lambda x: x.round(0))
df['price']= df[['price']].apply(lambda x: x.round(2))
df
```

```python
df.to_csv('listings-filtered.csv')
```
